```python
import sys
import math
import copy

#Clase sala 

class sala:
    def __init__(self, estado, camino):
        self.estado = estado
        self.camino = camino
        self.coste = len(camino)
        self.heuristica = 0
        self.ch = self.coste + self.heuristica
        self.tipo = tipoSala(tablero,self.estado[0],self.estado[1])
        self.entrada = 'TOP'
        self.salida = salas[abs(self.tipo)][self.entrada]
    def set_heuristica(self, heuristica):
        self.heuristica = heuristica
        self.ch = self.coste + self.heuristica
    def actualiza_coste(self):
        self.coste = len(self.camino)
    def actualiza_entrada(self,posi):
        self.entrada = posi
    def actualiza_tipo(self,tipo):
        self.tipo = tipo
    def actualiza_salida(self):
        self.salida = salas[self.tipo][self.entrada]
        
#Diccionarios para facilitar la b√∫squeda

salas = {0:  {},
         1:  {'TOP': 'DOWN','LEFT': 'DOWN','RIGHT': 'DOWN'},
         2:  {'LEFT': 'RIGHT','RIGHT': 'LEFT'},
         3:  {'TOP': 'DOWN'},
         4:  {'TOP': 'LEFT','RIGHT': 'DOWN'},
         5:  {'TOP': 'RIGHT','LEFT': 'DOWN'},
         6:  {'LEFT': 'RIGHT','RIGHT': 'LEFT'},
         7:  {'TOP': 'DOWN','RIGHT': 'DOWN'},
         8:  {'LEFT': 'DOWN','RIGHT': 'DOWN'},
         9:  {'TOP': 'DOWN','LEFT': 'DOWN'},
         10: {'TOP':'LEFT'},
         11: {'TOP': 'RIGHT'},
         12: {'RIGHT': 'DOWN'},
         13: {'LEFT': 'DOWN'},
         -1:  {'TOP': 'DOWN','LEFT': 'DOWN','RIGHT': 'DOWN'},
         -2:  {'LEFT': 'RIGHT','RIGHT': 'LEFT'},
         -3:  {'TOP': 'DOWN'},
         -4:  {'TOP': 'LEFT','RIGHT': 'DOWN'},
         -5:  {'TOP': 'RIGHT','LEFT': 'DOWN'},
         -6:  {'LEFT': 'RIGHT','RIGHT': 'LEFT'},
         -7:  {'TOP': 'DOWN','RIGHT': 'DOWN'},
         -8:  {'LEFT': 'DOWN','RIGHT': 'DOWN'},
         -9:  {'TOP': 'DOWN','LEFT': 'DOWN'},
         -10: {'TOP':'LEFT'},
         -11: {'TOP': 'RIGHT'},
         -12: {'RIGHT': 'DOWN'},
         -13: {'LEFT': 'DOWN'}}
salasIzq = {1:1,
            2:3,
            3:2,
            4:5,
            5:4,
            6:9,
            7:6,
            8:7,
            9:8,
            10:13,
            11:10,
            12:11,
            13:11}

salasDer = {1:1,
            2:3,
            3:2,
            4:5,
            5:4,
            6:7,
            7:8,
            8:9,
            9:6,
            10:11,
            11:12,
            12:13,
            13:10}
entradaSiguienteSala = {'DOWN': 'TOP',
                        'LEFT': 'RIGHT',
                        'RIGHT': 'LEFT'}

def tipoSala(m, x, y):  #Te devuelve el tipo de sala
    return(m[y][x])

def esfinal(sala, estado_final):  #Comprueba si el estado es el final
    return sala.estado == estado_final

def siguienteSala(sala,nuevaSala): #Te devuelve la posicion de indy en la siguiente sala
    if(sala.salida == 'DOWN'):
        nuevaSala.estado[1] += 1
    elif(sala.salida == 'RIGHT'):
        nuevaSala.estado[0] += 1
    elif(sala.salida == 'LEFT'):
        nuevaSala.estado[0] -= 1
        
            
def estaen(lista, sala): #Empleado en el algoritmo en anchura
    for i in lista:
        if i.estado == sala.estado:
            return True
    return False

def girarIzq(sala): #Rota la siguiente sala a la izquierda
    sucesor = copy.deepcopy(sala)
    siguienteSala(sala,sucesor)
    sucesor.actualiza_tipo(tipoSala(tablero,sucesor.estado[0],sucesor.estado[1]))
    sucesor.actualiza_tipo(salasIzq[sucesor.tipo])
    sucesor.actualiza_entrada(entradaSiguienteSala[sala.salida])
    if not esPosibleEntrar(sala,sucesor):
        return None
    sucesor.actualiza_salida()
    sucesor.camino.append("{} {} LEFT".format(sucesor.estado[0],sucesor.estado[1]))
    return sucesor

def girarDer(sala): #Rota la siguiente sala a la derecha
    sucesor = copy.deepcopy(sala)
    siguienteSala(sala,sucesor)
    sucesor.actualiza_tipo(tipoSala(tablero,sucesor.estado[0],sucesor.estado[1]))
    sucesor.actualiza_tipo(salasDer[sucesor.tipo])
    sucesor.actualiza_entrada(entradaSiguienteSala[sala.salida])
    if not esPosibleEntrar(sala,sucesor):
        return None
    sucesor.actualiza_salida()
    sucesor.camino.append("{} {} RIGHT".format(sucesor.estado[0],sucesor.estado[1]))
    return sucesor

def esperar(sala): #No rota ninguna sala
    sucesor = copy.deepcopy(sala)
    siguienteSala(sala,sucesor)
    sucesor.actualiza_tipo(tipoSala(tablero,sucesor.estado[0],sucesor.estado[1]))
    sucesor.actualiza_entrada(entradaSiguienteSala[sala.salida])
    sucesor.actualiza_salida()
    sucesor.camino.append("WAIT")
    return sucesor
    
def esPosibleEntrar(sala,sucesor): #Mira si es posible entrar en la sala
    if entradaSiguienteSala[sala.salida] not in salas[sucesor.tipo]:
        return False
    return True

def esPosibleGirar(sala): #Mira si es posible girar la sala(si el tipo es negativo no puede girarse)
    if (sala.tipo <= 0):
        return False
    return True

def conseguirSucesores(sala): #Funcion que devuelve los posibles sucesores
    posibles = []
    prueba = copy.deepcopy(sala)
    siguienteSala(sala,prueba)
    prueba.actualiza_tipo(tipoSala(tablero,prueba.estado[0],prueba.estado[1]))
    prueba.actualiza_entrada(entradaSiguienteSala[sala.salida])
    if prueba.tipo == 4:
        aux = girarDer(sala)
        posibles.append(aux)
    if esPosibleEntrar(sala,prueba):
        aux = esperar(sala)
        posibles.append(aux)
    else:
        aux1 = girarDer(sala)
        aux2 = girarIzq(sala)
        posibles.append(aux1)
        posibles.append(aux2)
    return posibles

def esperarRocas(sala,k): #Empleado en el mapa tipo 1 de rocas
    sucesor = copy.deepcopy(sala)
    siguienteSala(sala,sucesor)
    sucesor.actualiza_tipo(tipoSala(tablero,sucesor.estado[0],sucesor.estado[1]))
    sucesor.actualiza_entrada(entradaSiguienteSala[sala.salida])
    sucesor.actualiza_salida()
    if k == 0:
        sucesor.camino.append("{} {} RIGHT".format(3,2))
        k+=1
    elif k == 1:
        sucesor.camino.append("{} {} RIGHT".format(3,3))
        k+=1
    elif k == 2:
        sucesor.camino.append("{} {} RIGHT".format(3,4))
        k+=1
    elif k == 3:
        sucesor.camino.append("{} {} RIGHT".format(3,6))
        k+=1
    else:
        sucesor.camino.append("WAIT")
    return sucesor,k

def conseguirSucesoresRocas(sala,k): #Empleado en el mapa tipo 1 de rocas
    posibles = []
    prueba = copy.deepcopy(sala)
    siguienteSala(sala,prueba)
    prueba.actualiza_tipo(tipoSala(tablero,prueba.estado[0],prueba.estado[1]))
    prueba.actualiza_entrada(entradaSiguienteSala[sala.salida])
    if prueba.tipo == 11:
        aux = girarIzq(sala)
        posibles.append(aux)
    elif esPosibleEntrar(sala,prueba):
        aux,k = esperarRocas(sala,k)
        posibles.append(aux)
    else:
        aux1 = girarDer(sala)
        aux2 = girarIzq(sala)
        posibles.append(aux1)
        posibles.append(aux2)
    return posibles,k

def esperarRocas2(sala,k): #Empleado en el mapa tipo 2 de rocas
    sucesor = copy.deepcopy(sala)
    siguienteSala(sala,sucesor)
    sucesor.actualiza_tipo(tipoSala(tablero,sucesor.estado[0],sucesor.estado[1]))
    sucesor.actualiza_entrada(entradaSiguienteSala[sala.salida])
    sucesor.actualiza_salida()
    if k == 0:
        sucesor.camino.append("{} {} RIGHT".format(6,2))
        k+=1
    elif k == 1:
        sucesor.camino.append("{} {} RIGHT".format(7,3))
        k+=1
    elif k == 2:
        sucesor.camino.append("{} {} RIGHT".format(6,4))
        k+=1
    elif k == 3:
        sucesor.camino.append("{} {} RIGHT".format(6,6))
        k += 1
    else:
        sucesor.camino.append("WAIT")
    return sucesor,k

def conseguirSucesoresRocas2(sala,k):#Empleado en el mapa tipo 2 de rocas
    posibles = []
    prueba = copy.deepcopy(sala)
    siguienteSala(sala,prueba)
    prueba.actualiza_tipo(tipoSala(tablero,prueba.estado[0],prueba.estado[1]))
    prueba.actualiza_entrada(entradaSiguienteSala[sala.salida])
    if prueba.tipo == 11:
        aux = girarIzq(sala)
        posibles.append(aux)
    elif esPosibleEntrar(sala,prueba):
        aux,k = esperarRocas2(sala,k)
        posibles.append(aux)
    else:
        aux1 = girarDer(sala)
        aux2 = girarIzq(sala)
        posibles.append(aux1)
        posibles.append(aux2)
    return posibles,k

########## Busqueda en anchura
##Inicializacion

w, h = [int(i) for i in input().split()]
tablero = []
for i in range(h):
    line = input()  # each line represents a line in the grid and contains W integers T. The absolute value of T specifies the type of the room. If T is negative, the room cannot be rotated.
    tablero.append(list(map(int,line.split())))
ex = int(input())  # the coordinate along the X axis of the exit.
inputs = input().split()
xi = int(inputs[0])
yi = int(inputs[1])
posi = inputs[2]
p = (w==10 and h==8) #Comprobaciones
q = (w==9 and h==8) #Comprobaciones
z = (w==9 and h==9) #Comprobacioens
r = int(input())  # the number of rocks currently in the grid.
for i in range(r):
    inputs = input().split()
    xr = int(inputs[0])
    yr = int(inputs[1])
    posr = inputs[2]
estado_inicial = [xi,yi,salas[(tablero[yi][xi])][posi]] #Estado inicial siempre
estado_final = [ex,h-1,'DOWN'] #Este es el estado final siempre
inicial = sala(estado_inicial, [])
abiertos = []
abiertos.append(inicial)
cerrados = []
solucion = []
k = 0
j = 0
while (len(abiertos) > 0):
    actual = abiertos[0]
    abiertos.pop(0)
    if j != 0:
        print(actual.camino[0])
        actual.camino.pop(0)
    cerrados.append(actual)
    j+=1
    if esfinal(actual,estado_final):
        print("Hay solucion")
        print(actual.estado)
        break
    else:
        if (p and tablero[0][1] == -3) or z: #Segundo tipo mapa rocas
            sucesoreslista,k = conseguirSucesoresRocas2(actual,k)
            for i in sucesoreslista:
                if i is None:
                    continue
                if (estaen(abiertos,i) == False and estaen(cerrados,i) == False):
                    abiertos.append(i)
        elif p or q: #Primer tipo mapa rocas
            sucesoreslista,k = conseguirSucesoresRocas(actual,k)
            for i in sucesoreslista:
                if i is None:
                    continue
                if (estaen(abiertos,i) == False and estaen(cerrados,i) == False):
                    abiertos.append(i)
        else: #Primeros mapas
            sucesoreslista = conseguirSucesores(actual)
            for i in sucesoreslista:
                if i is None:
                    continue
                if (estaen(abiertos,i) == False and estaen(cerrados,i) == False):
                    abiertos.append(i)
```
